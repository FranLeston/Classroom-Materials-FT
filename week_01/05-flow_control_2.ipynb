##For

Es muy habitual escribir un c칩digo con bucles dentro de estas dos posibles situaciones:    
    - 1: Cuando utilizamos un contador en el seno de una repetici칩n. El contador es una variable que va increment치ndose o disminuyendo de forma constante en cada iteraci칩n del bucle hasta alcanzar un valor l칤mite que marca el fin de las repeticiones.   
    - 2: Cuando iteramos sobre los elementos de un contenedor, por ejemplo, una lista, para operar sobre cada uno de ellos. Tambi칠n en este caso conocemos de antemano el n칰mero de repeticiones que tendr치n lugar.

La sentencia For, en Python, permite iterar sobre secuencias de valores (listas o cadenas) seg칰n un orden establecido. A diferencia de otros lenguajes, con el for de Python no se definen contadores o expresiones que cambien a cada iteraci칩n sino una secuencia sobre la cual la variable contador tomar치 sus valores. En cualquier caso, es la opci칩n m치s adecuada para resolver los dos casos anteriores....vamos a poner un ejemplo:

`````python
for num in [1,2,3]:
    print(num * 2)
`````

La sentencia for, en Python, itera sobre colecciones. Por tanto, si necesitamos un contador que se incremente en cada iteraci칩n, Python proporciona la funci칩n range(), que genera listas de enteros. Esta funci칩n permite generar secuencias y es bastante vers치til y podemos invocarla de diferentes maneras:

- Con un solo argumento: genera enteros que van desde 0 hasta el anterior indicado como par치metro:

`````python
print(list(range(5)))
`````

`````python
n = 5
`````

`````python
print(list(range(n+1)))
`````

- Con dos argumentos. El primer par치metro es el valor inicial y el segundo el valor por debajo del cual deben estar los elementos de la lista:

`````python
print(list(range(5,10)))
`````

- Con tres argumentos. Igual que en el anterior, pero el tercer par치metro indica el incremento que se produce de un elemento al siguiente:

`````python
print(list(range(0,11,2)))
`````

`````python
for num in range(11):
    print(num*5)
`````

### Iterate over a list

`````python
nums = [1,2,3,4]
nums_multiplicados = []
for num in nums:
    nums_multiplicados.append(num*3)
print(nums_multiplicados)
`````

###Iterate over lists with their index

`````python
nombres = ["Fer", "Ana", "Sheriff", "Ras"]
for indice,nombre in enumerate(nombres):
    print(f"El nombre es {nombre} y su posici칩n es {indice}")
`````

### Iterate over two lists at the same time

`````python
nombres = ["Fer", "Ana", "Sheriff", "Ras","Fran"]
emojis = ["游댠", "游꺖", "游놅","游"]
lista_3 = [1,2,3,4,4,5,6,7,7,8]
`````

`````python
for nombre,emoji,num in zip(nombres,emojis,lista_3):
    print(f"Hola {nombre} ---> {emoji}, {num}")
`````

`````python
#Enumerate + zip a mano
`````

`````python
tuplas = []
for nombre,emoji in zip(nombres,emojis):
    tuplas.append((nombre,emoji))
print(tuplas)
`````

`````python
for indice,elemento in enumerate(tuplas):
    print(f"El elemento es {elemento} y el 칤ndice es {indice}")
`````

`````python
#Enumerate + zip en el bucle directamente
`````

`````python
for indice,(nombre,emoji) in enumerate(zip(nombres,emojis)):
    print (f"El nombre es {nombre} su emoji es {emoji} y su 칤ndice en la lista es {indice}")
`````

##Break

We can literally `break` a recursion when a condition is satisfied

`````python
musicians = ["Rufus T. Firefly", "Viva Suecia", "Standstill","Bowie", "Fuel Fandango"]
`````

`````python
for group in musicians:
    print(f"{group} hace temazos")
    if len(group)<=5:
        break
`````
